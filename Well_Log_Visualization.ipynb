```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
log_df = pd.read_csv('log.csv')
```

```python
log_df
```

### Columns description:

- Depth: Depth at which the log parameters are taken
- RxoRt: Resistivity log reading
- RLL3
- SP: Spontaneous Potential log reading
- RILD: Deep resistivity log reading
- MN
- MI
- MCAL
- DCAL
- RHOB: Density log reading
- RHOC
- DPOR
- CNLS
- GR: gamma ray log reading

```python
plot_columns=log_df.columns[1:]
fig,axes = plt.subplots(1,len(plot_columns), figsize=(30,40))
for i in range(len(plot_columns)):
#      col=(np.random.random(),np.random.random(),np.random.random())
     colors=['red', 'black', 'blue', 'purple', 'green', 'violet', 'Teal']
     axes[i].plot(log_df[plot_columns[i]],log_df.Depth,c=random.choice(colors))
     axes[i].set_title(plot_columns[i], fontname='Times New Roman', size =28, fontweight='bold');
```
